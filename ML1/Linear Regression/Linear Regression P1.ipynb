## Types of Unconstrained minimisation

In the previous segment, you understood the concept of cost function and how can it be minimised and while minimising the sum of squares we differentiated w.r.t 'm' and 'c' to find out their values. In that case, there were no restrictions on the values the function parameters - 'm' and 'c'. This is known as unconstrained minimisation. Consider a situation where there is a constraint i.e. a condition on 'm' or 'c' such as m>2 or c < -1 etc. How will this affect the results? 


**You learnt about two types of optimisation: constrained and unconstrained optimisation.** You also discussed that there are two ways to go about unconstrained optimisation: differentiation and using gradient descent. You saw the example of optimisation using differentiation while coming up with the trend line. Gradient descent in detail is covered in the main module, but before that, let get an intuition behind how gradient descent works.


So there are two types of minimisations, namely constrained and unconstrained minimisations.  
unconstrained minimisation can be solved using two methods:

**Closed form method:** The function to be minimised is simply differentiated and equated to 0 to achieve a solution. The solution is also double differentiated to check if the solution is greater than 0.

**Gradient Descent:** It is an iterative minimisation method which reaches the minima step by step (as shown in the figure below).You start with an initial assumed value of the parameter. This initial assumed value can be anything (say $X_{0}$). Then you assume $\alpha$ which is rate of learning. For that value ($X_{0}$), you calculate the output of the differentiated function which we denote as $f^{'}(x)$. Then the new value of the parameter becomes $x -f^{'}(x)*\alpha$ . You continue the process untill the algorithm reaches an optimum point ($X_{4}$); i.e the value of the parameter does not change effectively after this point.

# Gradient Descent-Linear Regression


>**Gradient descent is an optimization algorithm used to find the values of the parameters (coefficients) of a function (f) that minimizes a given cost function (cost).**

**Linear Regression Cost Function**
equation for the line that's fit the data, is given as:
   
   $$ y(p)= \beta_{0} + \beta_{1}x $$
   
Where $\beta_{0}$ is the intercept of the fitted line and $\beta_{1}$ is the coefficient for the independent variable x.

>So first step is to find the cost function and in second step we will try to find out the minimum value of cost function which is nothing but optimum value of $\beta_{0}$ and $\beta_{1}$.
To minimise the cost function, we will use optimization algorithmns such as **gradient descent**


To summarise, in this section, we understood that to find the optimum betas, we need to reduce the cost function for all data points, which is given as,

>$$ J(\theta_{0},\theta_{1}) = \sum_{i=1}^N (y_{i} - y_{i}(p))^2$$
 

**The way to find the optimal betas or thetas is know as Gradient Descent.** 

## How To minimise the cost function 
Answer is optimisation methods

### Optimisation Methods

There are two types of Optimisation methods,

> * Closed form solution
* Iterative form solution
  * First order method ( Gradient Descent)  | in this we compute $\frac{d(J(\theta)}{d(\theta)}$ )
  * Second order method ( Newton's method)  | In this we compute second order differentials 
$\frac{d^2 J(\theta)}{d \theta^2}$


So, Gradient descent is an iterative form solution of order one. So to compute optimal thetas, we need to apply Gradient Descent to the Cost function, which is given as follows,
$$\frac{d J(\theta)}{d(\theta)}$$

**Closed Form**

let's say we have a function $\theta^2$ (parabola). So in closed form we can solve the value of minima as below:

$J(\theta) = \theta^2$

$\frac{d J(\theta)}{d(\theta)} = 2 \theta$

$J'(\theta)=0$

$2\theta =0$

$\theta = 0$   which implies we have minima at $\theta =0 $ for the function $\theta^2$ 


**Iterative method (Gradient Descent)**

In this we will start from a point lets say $\theta =10 $

$\theta^0 =10$

$\theta^1 =\theta^0 - \eta \frac{d J(\theta)}{d(\theta)}$ (at $\theta = \theta^0$)   

$\frac{d J(\theta)}{d(\theta)} = 2 \theta$ at $\theta =10 = 20$

here $\eta$ is learning rate and for time being let's take this value as 0.1

$\theta^1 =10 - 0.1* 20 = 8.0$

we have started at 10 and we had moved in the direction where the cost function is becoming lesser and lesser.
and we have moved to 8 similarly if we keep taking more steps we will start to reach the solution


>Gradient descent is an iterative method of optimising an objective function, in our case the cost function, by moving toward the negative of the gradient.
To compute $ \theta{1}$ ,we saw that the equation will look like this,
$$\theta^1 =\theta^0 - \eta \frac{d J(\theta)}{d(\theta)}$$

>Generic equation: $$\theta^t =\theta^t-1 - \eta \frac{d J(\theta)}{d(\theta)}  |at   \theta = \theta^t-1$$


**Where $\eta$ is known as the learning rate, which defines the speed at which we want to move towards negative of the gradient.**


## Learning Rate

In the previous lecture, we learned that it's easy to go for a closed form solution than using an iterative form solution (gradient descent) to optimise the cost function.
But the example we looked was a simple 1D example, what will happen if the cost function is in higher dimensions?
To summarise, though gradient descent looks complicated for a 1D function, it's easier to compute the optimal minima using gradient descent for higher dimension function. We will be using it in logistic regression and even for neural networks. 


To summarise, a large value of learning rate may oscillate your solution, and you may skip the optimal solution (global minima). So it's always a good practice to choose a small value of learning rate and slowly move towards the negative of the gradient.

**What will happen if we take large value of learning rate:**
* oscillation issue

**what will happen if we take very small value of learning rate:**
* slow convergence
